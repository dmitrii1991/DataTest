```python
# Подготовка
import numpy as np
import pandas as pd
```

```python
######################################################################## 
# 1 Загрузить БД данных о транзакциях по сделкам с недвижимостью на сервер. 
######################################################################## 
data = pd.read_csv('transactions.csv', delimiter=',')

#  исправление ошибок с датами 
data.at[313684, "instance_date"] = '23-11-2022' # было 23-11-1422
data.at[373305, "instance_date"] = '02-07-2016' # было 02-07-1416
data.at[1040345, "instance_date"] = '04-02-2017' # было 04-02-1417
data.at[1102031, "instance_date"] = '30-01-2020' # было 30-01-1420
data
```

```python
######################################################################################################################## 
# 2 Сделать выборку за 2023 г. по продажам коммерческой недвижимости в таблицу следующего вида по выделенным признакам  
###################################################################################################################### 
dataformat = data[['transaction_id', 'trans_group_en', 'property_usage_en', 'procedure_name_en',
                  'property_sub_type_en', ## дубль
                  'instance_date', 'area_id', 'area_name_en', 'procedure_area', 'actual_worth', 'meter_sale_price']]
#преобразуем дату 
dataformat['instance_date']= pd.to_datetime(dataformat['instance_date'], format='%d-%m-%Y')
dataformat
```

```python
dataformat['trans_group_en'].unique() # виды сделок 
# Sales продажи
# Mortgages Ипотека
# Gifts подарки
```

```python
print(dataformat['transaction_id'].unique().size == dataformat['transaction_id'].size) # убеждаемся что дубликатов не)


```

```python
######################################################################################################################## 
# 3 Выгрузить результат сортировки в export_transactions_2023.csv для проверки с разделителем “,” и все значения в “...” (кавычках).
###################################################################################################################### 
import csv
filterdata = dataformat[(dataformat['instance_date'] >= '2023-01-01') & (dataformat['instance_date'] < '2024-01-01')]
filterdata = filterdata[filterdata['trans_group_en'] == 'Sales']
filterdata = filterdata[filterdata['property_usage_en'] == 'Commercial']
filterdata = filterdata[filterdata['procedure_name_en'] == 'Sell']

# проверки
print(filterdata['trans_group_en'].unique() == np.array(['Sales'], dtype=object))
print(filterdata['instance_date'].min() >= pd.Timestamp('2023-01-01 00:00:00'))
print(filterdata['instance_date'].max() <= pd.Timestamp('2023-12-31 00:00:00'))
filterdata.to_csv("export_transactions_2023.csv", sep=',', quoting=csv.QUOTE_ALL)
filterdata
```

```python
######################################################################################################################## 
# 4 Подсчитать общее количество сделок продаж за 2023 г. и общий валовый бюджет сделок по продажам коммерческой недвижимости за 2023 г.
# в долларах США по курсу на текущую дату - в простом текстовом виде (отчет).
###################################################################################################################### 

# выгружаем БД по курсу
# https://ru.investing.com/currencies/aed-usd-historical-data
AED_USD = pd.read_csv('AED_USD.csv', delimiter=',')
AED_USD['Дата']= pd.to_datetime(AED_USD['Дата'], format='%d.%m.%Y')
AED_USD
```

```python
# получаем цену курса на дату тек
AED_USD=AED_USD[['Дата', 'Цена']]
dict_AED_USD =  AED_USD.set_index('Дата').T.to_dict('index')
# dict_AED_USD['2023-01-30']
dict_AED_USD
```

```python
def get(key: pd.Timestamp):
    while True:
        try:
            return float(dict_AED_USD["Цена"][key].replace(',', '.'))
        except:
           key -= pd.Timedelta(days=1) 

def sum_price(dataframe):
    amount_USD = 0
    amount_AED = 0
    for index, row in dataframe.iterrows():
        amount_USD += get(row["instance_date"]) * row["actual_worth"]
        amount_AED += row["actual_worth"]
    return amount_USD, amount_AED
amount_USD, amount_AED = sum_price(filterdata)
(amount_USD, amount_AED)
```

```python

amount = filterdata["transaction_id"].size
f"Total {amount} commercial deals per 01.01.2023 till 31.12.2023 in total amount AED {amount_AED} = $ {amount_USD} = $ {amount_USD/1000_000_000:.2f} млрд. *"
```

```python
######################################################################################################################## 
# 5 Затем выгрузить сводки аналогично пункту 4 по кварталам 2023 г. в следующем виде.
# Таблицу сохранить с названием “deals-by-quarters-2023.csv”.
###################################################################################################################### 
```

```python
filterdata1 = filterdata[(filterdata['instance_date'] >= '2023-01-01') & (dataformat['instance_date'] < '2023-04-01')]

filterdata2 = filterdata[(filterdata['instance_date'] >= '2023-04-01') & (dataformat['instance_date'] < '2023-07-01')]

filterdata3 = filterdata[(filterdata['instance_date'] >= '2023-07-01') & (dataformat['instance_date'] < '2023-10-01')]

filterdata4 = filterdata[(filterdata['instance_date'] >= '2023-10-01') & (dataformat['instance_date'] < '2024-01-01')]
```

```python

details = {
    'Quarter I, 2023' : [filterdata1["transaction_id"].size, sum_price(filterdata1)[0], sum_price(filterdata1)[0] / filterdata1["transaction_id"].size],
    'Quarter II, 2023' : [filterdata2["transaction_id"].size, sum_price(filterdata2)[0], sum_price(filterdata2)[0] / filterdata2["transaction_id"].size],
    'Quarter III, 2023' : [filterdata3["transaction_id"].size, sum_price(filterdata3)[0], sum_price(filterdata3)[0] / filterdata3["transaction_id"].size],
    'Quarter IV, 2023' : [filterdata4["transaction_id"].size, sum_price(filterdata4)[0], sum_price(filterdata4)[0] / filterdata4["transaction_id"].size],
}

df = pd.DataFrame(details, index = ['Quantity of sales in commercial propertyDubai (UAE)', 'Total sum of sales per quarter 2023', 'Median sum of sales per quarter 2023'])
pd.options.display.float_format ='{:,.0f}'.format
df.to_csv("deals-by-quarters-2023.csv", sep=',', quoting=csv.QUOTE_ALL)

amount_ = 0
totalprice_ = 0

for key, value in details.items():
     amount_ += value[0]
     totalprice_ += value[1]
print(f"{amount}=={amount_} {amount_ == amount}")
print(f"{amount_dollars}=={totalprice_} {totalprice_ == amount_dollars}")

df
```

```python
######################################################################################################################## 
# Выгрузить в таблицу по кварталам  2023 года количество сделок и совокупный бюджет по каждому типу коммерческой недвижимости 
# и общую сумму по каждому виду коммерческой недвижимости
###################################################################################################################### 


property_usage_types = list(filterdata["property_sub_type_en"].unique())
property_usage_types.remove(np.nan)
```

```python
details2 = {
    'Total deals by property usage type' : [],
    'Median cost per sq meter by property usage type' : [],
    'Quarter I, 2023' : [],
    'Quarter II, 2023' : [],
    'Quarter III, 2023' : [],
    'Quarter IV, 2023' : [],
}


for property_sub_type_en in property_usage_types:
    for key in details2.keys():
        match key:
            case  'Total deals by property usage type':
                 details2[key].append(filterdata[filterdata['property_sub_type_en'] == property_sub_type_en]["transaction_id"].size)
            case 'Median cost per sq meter by property usage type':
                 details2[key].append(filterdata[filterdata['property_sub_type_en'] == property_sub_type_en]["meter_sale_price"].sum() / details2['Total deals by property usage type'][-1])
    
            case 'Quarter I, 2023':
                details2[key].append(filterdata1[filterdata1["property_sub_type_en"] == property_sub_type_en]["transaction_id"].size)
            case 'Quarter II, 2023':
                 details2[key].append(filterdata2[filterdata2['property_sub_type_en'] == property_sub_type_en]["transaction_id"].size)
            case 'Quarter III, 2023':
                 details2[key].append(filterdata3[filterdata3['property_sub_type_en'] == property_sub_type_en]["transaction_id"].size)
            case 'Quarter IV, 2023':
                 details2[key].append(filterdata4[filterdata4['property_sub_type_en'] == property_sub_type_en]["transaction_id"].size)

for key in details2.keys():
    match key:
        case  'Total deals by property usage type':
             details2[key].append(filterdata[filterdata['property_sub_type_en'].isna()]["transaction_id"].size)
        case 'Median cost per sq meter by property usage type':
             details2[key].append(filterdata[filterdata['property_sub_type_en'].isna()]["meter_sale_price"].sum() / details2['Total deals by property usage type'][-1])
        # case 'Median cost per sq meter by property usage type':
        #      details2[key].append(
        #          sum_price_by_meter(filterdata[filterdata['property_sub_type_en'] == property_sub_type_en]) 
        #          / details2['Total deals by property usage type'][-1])
    
        case 'Quarter I, 2023':
            details2[key].append(filterdata1[filterdata1["property_sub_type_en"].isna()]["transaction_id"].size)
        case 'Quarter II, 2023':
             details2[key].append(filterdata2[filterdata2['property_sub_type_en'].isna()]["transaction_id"].size)
        case 'Quarter III, 2023':
             details2[key].append(filterdata3[filterdata3['property_sub_type_en'].isna()]["transaction_id"].size)
        case 'Quarter IV, 2023':
             details2[key].append(filterdata4[filterdata4['property_sub_type_en'].isna()]["transaction_id"].size)

property_usage_types2 =list(property_usage_types)
property_usage_types2.append(np.nan)

df2= pd.DataFrame(details2, index = list(property_usage_types2))
df2.to_csv("deals-by-property-usage-type.csv", sep=',', quoting=csv.QUOTE_ALL)
df2
```
